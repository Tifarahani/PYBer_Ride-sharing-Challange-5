```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# Create label for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker= "*", color="Red", linewidth=2, label="Boston")
# Create label for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create plot with object oriented approach
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
```

```python
# Create the bar chart.
plt.bar(x_axis, y_axis)
```

```python
# Create the bar chart.
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fares($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal bar chart.
plt.barh(x_axis, y_axis)
```

```python
# Create the pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Refine the pie chart
plt.subplots(figsize=(8, 8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%.1f%%")
plt.show()
```

```python
# Create a pie chart with object oriented method
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green","yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, 
       autopct="%.1f%%", 
       explode=explode_values, 
       colors=colors, 
       shadow=True,
       startangle=90,
       counterclock=False)
plt.show()
```
